```python
# Plot a line graph of the happiest scores in each year to find the happiest country in the world

P2019 = y2019.head()
P2018 = y2018.head()
P2017 = y2017.head()
P2016 = y2016.head()
P2015 = y2015.head()

Country19 = P2019 ["Country or region"]
Happiness19 = P2019 ["Score"]

Country18 = P2018 ["Country or region"]
Happiness18 = P2018 ["Score"]

Country17 = P2017 ["Country"]
Happiness17 = P2017 ["Happiness.Score"]

Country16 = P2016 ["Country"]
Happiness16 = P2016 ["Happiness Score"]

Country15 = P2015 ["Country"]
Happiness15 = P2015 ["Happiness Score"]

  
plt.plot(Country19, Happiness19, label = "2019")
plt.plot(Country18, Happiness18, label = "2018")
plt.plot(Country17, Happiness17, label = "2017")
plt.plot(Country16, Happiness16, label = "2016")
plt.plot(Country15, Happiness15, label = "2015")

plt.title("Happiness of Different Countries in 2019")
plt.xlabel("Country")
plt.ylabel("Score")
plt.legend()
plt.show()

print("Finland is the happiest country in the World with a score of 7.769 in 2019")
```

```python
# Plot a trendline to see the growth ot decline of happiness globally over the years (2015 - 2019)

T2019 = y2019["Score"].mean()
T2018 = y2018["Score"].mean()
T2017 = y2017["Happiness.Score"].mean()
T2016 = y2016["Happiness Score"].mean()
T2015 = y2015["Happiness Score"].mean()

trendline = {"Year":["2015", "2016", "2017", "2018", "2019"],
        "Score":[T2015, T2016, T2017, T2018, T2019]}
 
Trendline = pd.DataFrame(trendline)

Year = Trendline["Year"]
Score = Trendline["Score"]

plt.plot(Year, Score)

plt.title("Trendline of Happiness")
plt.xlabel("Year")
plt.ylabel("Score")
plt.show()

print ("Overall Happiness increased over the years")
```
